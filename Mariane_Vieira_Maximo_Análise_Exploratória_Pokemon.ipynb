## **Tipos dos pokemons**

```python
#tabela com os tipos primários e secundários dos pokemons

tipos_ps = dados[['name','type1','type2']]
tipos_ps
```

```python
#analisando quais são os tipos de pokemons na tabela
tipos = dados['type1'].unique()
tipos
```

```python
eixoy = dados['type1'].value_counts()

cinza_claro = "#CFCDCE"
cinza_medio = "#A2A1A3"
cinza_escuro = "#787878"
azul_claro = "#8db7eb"
azul_escuro = "#253760"

cores = [cinza_medio, cinza_escuro, azul_escuro]
```

```python
fig, ax = plt.subplots(figsize = (15,4))
ax.bar(x = tipos, height = eixoy, color = cores )

#remover grids e eixos 
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

#título personalizado
x_text_antes = 0.2

plt.text(x = x_text_antes,
         y = 130,
         s = 'Distribuição dos Pokemons por tipo primário',
         fontsize = 14,
         color = cinza_escuro,
         weight = 'bold');

##legenda
#ax.set_xsticks(x)
#ax.set_xticklabels(labels)

#texto
plt.text(x = 5.5,
          y = 110,
          s ="Grande parte dos  \n pokemons são \n do tipo $\\bf(Grass)$.\n.",
          verticalalignment = "top",
          fontsize = 12,
          color = cinza_medio);


```

```python
#pokemons com os tipos primários Grass
grass = tipos_ps[tipos_ps['type1'] == 'grass']
grass
```

```python
##pokemons fracos contra o tipo Grass ( no jogo pokemons que possuem a fraquesa < 1 são fracos ao elemento )

#primeiro vamos separar a tabela só com estes dados 

dano = dados[['name', 'against_bug',	'against_dark',	'against_dragon',	'against_electric',	'against_fairy',	'against_fight',	'against_fire',	'against_flying',	'against_ghost',	'against_grass',	'against_ground',	'against_ice',	'against_normal',	'against_poison',	'against_psychic',	'against_rock',	'against_steel',	'against_water']
]

#agora vamos vizualizar somente a coluna referente aos danos causados pelo Fire

dano_grass = dados[['name', 'against_grass']]
fracos = dano_grass[dano_grass['against_grass'] < 1.0]
fracos
```

## **Pokemon mais Forte e mais Fraco**

```python
#para determinar se um pokemon é forte vemos o valor do seu IV que é uma junção do poder de ataque e defesa
#para os pokemons mais fortes temos:

IV = dados[['name', 'type1', 'sp_attack', 'sp_defense']]
pok_forte = IV[(IV['sp_attack'] >= 120) & (IV['sp_defense'] >= 120)]
pok_forte
```

```python
#para os pokemons mais fracos temos:

IV = dados[['name', 'type1', 'sp_attack', 'sp_defense']]
pok_fraco = IV[(IV['sp_attack'] <= 20) & (IV['sp_defense'] <= 40)]
pok_fraco
```

```python
dados['percentage_male']
```

```python
#verificando todas as possíveis porcentagens de pokemon macho na natureza

dados['percentage_male'].unique()
```

```python
#para melhor vizualização: gráfico da proporção de pokemons machos na natureza
#gráfico da proporção de pokemons machos na natureza

dados_percentage = dados['percentage_male'].value_counts()

a = dados_percentage.loc[0]
b = dados_percentage.loc[11.2]
c = dados_percentage.loc[24.6]
d = dados_percentage.loc[50]
e = dados_percentage.loc[75.4]
f = dados_percentage.loc[88.1]
g = dados_percentage.loc[100]

#construindo o gráfico
fig, ax = plt.subplots(figsize=(15,4))
ax.bar('0', a, width=0.55, color = cinza_claro)
ax.bar('11.2', b, width=0.55, color = azul_claro)
ax.bar('24.6', c, width=0.55, color = azul_escuro)
ax.bar('50', d, width=0.55, color = cinza_medio)
ax.bar('75.4', e, width=0.55, color = cinza_claro)
ax.bar('88.1', f, width=0.55, color = azul_claro)
ax.bar('100', g, width=0.55, color = azul_escuro)


#remover grids e eixos 
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

#título personalizado
x_text_antes = 0.2

plt.text(x = x_text_antes,
         y = 600,
         s = 'Distribuição das porcentagens de pokemons machos na natureza',
         fontsize = 14,
         color = cinza_escuro,
         weight = 'bold');

##legenda
#ax.set_xsticks(x)
#ax.set_xticklabels(labels)

#texto
plt.text(x = 0.5,
          y = 400,
          s =" Grande parte da  \n porcentagem de  \n pokemons machos \n está dividida em  $\\bf(50 \%)$.\n.",
          verticalalignment = "top",
          fontsize = 12,
          color = cinza_medio);
```

## **Pokemons Lendários**

```python
dados['is_legendary'].unique()
```

```python
#analisando a quantidade de pokemons legendários
dados['is_legendary'].value_counts()
```

```python
#para melhor vizualização: gráfico da proporção de pokemons legendários e não legendários


labels1 = 'Não lendário', 'Lendário'
colors = [cinza_claro, cinza_escuro]
legendario_plot = dados['is_legendary'].value_counts().unique()


fig1, ax1 = plt.subplots()
ax1.pie(legendario_plot, labels = labels1, autopct='%1.1f%%', colors = colors, shadow=True, startangle=360, explode=(0.2, 0.1))
ax1.axis('equal')

#título personalizado
x_text_antes = -2

plt.text(x = x_text_antes,
         y = 1.5,
         s = 'Relação entre pokemons Lendários e Não lendários',
         fontsize = 14,
         color = cinza_escuro,
         weight = 'bold');

plt.show()
```
